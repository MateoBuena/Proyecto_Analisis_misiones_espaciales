```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Space_Corrected.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['millions $']=pd.to_numeric(df[' Rocket'],errors='coerce')
df.head()
```

```python
mean=df['millions $'].mean()
df['millions $'].fillna(value=mean,inplace=True)
```

```python
df.head()
```

```python
df['millions $']=df['millions $'].astype(float).round(2)
```

```python
df.head()
```

```python
df.drop(' Rocket', axis=1, inplace=True)
df.columns
```

```python
df.drop('Unnamed: 0.1', axis=1, inplace=True)
df.columns
```

```python
df['Datum'].info()
```

```python
datum=df['Datum']
```

```python
datum
```

```python
datum.value_counts()
```

```python
dates=[]
for dat in datum:
    dat=pd.to_datetime(dat)
    dates.append(dat.year)

dates
```

```python
df['Year']=dates
```

```python
df.head()
```

```python
df.drop('Datum', axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Status Mission'].value_counts()
```

```python
df['Status Rocket'].value_counts()
```

```python
df['Location'].value_counts()
```

```python
location=df['Location']
```

```python
locations=[]
for loc in location:
    loc=loc.split(',')
    locations.append(loc[-1])
    
locations
```

```python
df['Country']=locations
```

```python
df.drop('Location',axis=1,inplace=True)
df.head()
```

```python
df['Country'].value_counts()
```

```python
df_especifico=df[df['Year']==2020]
```

```python
df['Year'].value_counts()
```

```python
sns.lineplot(df,x=df['Year'],y=df['millions $'])
plt.show()
```

```python
df.head()
```

```python
df_status_mission=df[df['Status Mission']=='Success'].groupby('Year').size()
```

```python
df_status_mission
```

```python
total_missions=df.groupby('Year').size()
total_missions
```

```python
porcentaje_missiones_exitosas=(df_status_mission/total_missions*100).fillna(0)
porcentaje_missiones_exitosas
```

```python
sns.lineplot(porcentaje_missiones_exitosas)
plt.ylabel('porcentaje de exito')
plt.show()
```

## se puede ver como el porcentaje de exito fue mejorando quedandose por encima del 90%

una conclusión que podemos sacar, es que, a medida que pasan los años, el coste de los lanzamientos es menor, y el porcentaje de exito sube o al menos se mantiene por encima del 80% desde los años 80
